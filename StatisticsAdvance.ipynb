```python
"""
Q1) Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis 
    testing?
Answer-
    The z-statistic, also known as the z-score, is a measure that describes the position of a data point relative to the mean of a dataset,
    expressed in terms of standard deviations. 

    The z-statistic for a sample mean is given by the formula:
    z = (sample mean) - (population mean) / standard error of the mean 
    where, standard error of the mean = population standard deviation divided by the square root of the sample size.

    While dealing with a single data point, the z-score would be: 
    z = (individual data point x) - mean of the population / standard deviation of the population.

    Relationship to the Standard Normal Distribution:
    The z-statistic is directly related to the standard normal distribution (also known as the Z-distribution), which has:
    mean of 0 and standard deviation of 1.
    A z-statistic indicates how many standard deviations a particular value is from the mean. Positive z-values indicate values above the mean,
    and negative z-values indicate values below the mean. The standard normal distribution is symmetrical, and the area under the curve 
    represents probabilities.

    Use in Hypothesis Testing:
    The z-statistic is a key tool in hypothesis testing which helps determine whether the observed data differs significantly from the null
    hypothesis.
    Steps:
    1. State the hypotheses:
       * Null hypothesis (H‚ÇÄ): Assumes no effect or no difference.
       * Alternative hypothesis (H‚ÇÅ): Assumes there is an effect or difference.
    
    2. Compute the z-statistic: Based on the sample data, calculate the z-statistic to measure how far the sample mean is from the population
       mean, assuming the null hypothesis is true.

    3. Find the critical value: This is determined by the significance level (ùõº), which typically is 0.05 for a 95% confidence level.
       Critical z-values correspond to the probability cutoffs (e.g., ¬±1.96 for a two-tailed test at Œ±=0.05).

    4. Compare the z-statistic to the critical value:
       * If the absolute value of the z-statistic is greater than the critical value, you reject the null hypothesis, indicating that the 
         sample provides enough evidence to conclude a significant difference.
       * If the z-statistic falls within the critical region, you fail to reject the null hypothesis, indicating insufficient evidence of a 
         significant difference.
"""
```

```python
"""
Q7) Explain the Central Limit Theorem (CLT) with example.
Answer-
    The Central Limit Theorem (CLT) says that if you take a large enough sample size from any population (no matter how it's shaped), the 
    average of those samples will form a normal (bell-shaped) distribution.
    Key Points:
    1. Shape Doesn't Matter: The population could be skewed, but the sample means will still look normal.
    2. Sample Size: A larger sample size (usually n ‚â• 30) makes this work better.
    3. Mean & Standard Error: The average of sample means will be close to the population mean, and the spread (standard error) depends on 
       the population standard deviation divided by the square root of the sample size.

    Example of the Central Limit Theorem - 
    Imagine we're interested in the average time people spend on a particular website each day. We don't know the actual distribution of 
    all user times; it might be heavily skewed since some people spend only a few seconds while others might spend hours.

    1. Population: Let's assume the average time spent is 20 minutes with a standard deviation of 10 minutes.
    2. Sampling: We take a sample of 50 people from this population and record the average time each person spent on the site.
    3. Distribution of Sample Means: If we repeat this process (taking new samples of 50 people and calculating each sample's mean time 
       spent), we get a distribution of sample means.
    4. Result of the CLT: According to the CLT, the distribution of these sample means will approximate a normal distribution with:
       * Mean Œº=20 minutes (same as the population mean), 
       * Standard Error œÉ / square root of n = 10 / square root of 50 ‚âà 1.41 minutes.
    
    So, even if the distribution of individual times is not normal, the distribution of the sample means will be approximately normal if the
    sample size is large (in this case, 50).
"""
```

```python
"""
Q8) Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be applied instead?
Answer-
    Z-scores and t-scores are both used in statistics to measure how far a sample mean or individual data point is from the population mean 
    in terms of standard deviations, but they are used in different situations:

    Z-Score -
    Definition: A z-score (or standard score) measures how many standard deviations a data point is from the mean.
    Formula: Z= (X - Œº) / œÉ
    where X is the data point, Œº is the population mean, and œÉ is the population standard deviation.
    
    Use a z-score when:
    1. The sample size is large (typically n>30).
    2. You know the population standard deviation (œÉ).
    3. The population follows a normal distribution or the sample size is large enough for the Central Limit Theorem to apply.

    T-Score
    Definition: A t-score is similar to a z-score but accounts for the uncertainty introduced by using a sample instead of the entire 
    population.
    Formula: T = (X -X bar) / (s / square root of n)
    where X is the sample mean, X bar is the sample mean, s is the sample standard deviation, and n is the sample size.
    
    Use a t-score when:
    1. The sample size is small (typically n<30).
    2. You don't know the population standard deviation and have to estimate it with the sample standard deviation.
    3. The data is approximately normally distributed (especially important for smaller sample sizes).

    Key Differences
    1. Sample Size: Z-scores are used for larger samples, t-scores for smaller samples.
    2. Population vs. Sample Standard Deviation: Z-scores require the population standard deviation, while t-scores use the sample standard
       deviation as an estimate.
    3. Distribution: Z-scores follow the standard normal distribution; t-scores follow the t-distribution, which is broader and has heavier 
       tails (accounts for more variability) when the sample size is small.
"""
```

```python
"""
Q9) Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and
    p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?
    Task: Write Python code to calculate the z-score and p-value for the given data.
    Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.
"""
from scipy.stats import norm

sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25
alpha = 0.05  

standard_error = std_dev / (sample_size ** 0.5)

z_score = (sample_mean - population_mean) / standard_error

p_value = 2 * (1 - norm.cdf(abs(z_score)))  # two-tailed p-value

# Determine the decision based on the p-value
decision = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

z_score, p_value, decision
```

```python
"""
Q10) Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. Generate 1,000 samples and plot the
     distribution. What is the expected mean and variance? 
     Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.
     Objective: Understand the properties of a binomial distribution and verify them through simulation.
Answer-
"""
import numpy as np
import matplotlib.pyplot as plt

n_trials = 10         
p_success = 0.6       
n_samples = 1000      

# Simulate binomial distribution
data = np.random.binomial(n=n_trials, p=p_success, size=n_samples)

# Calculate expected mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=range(n_trials + 2), align='left', color='skyblue', edgecolor='black', rwidth=0.8)
plt.title('Binomial Distribution (n=10, p=0.6) - 1,000 Samples')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n_trials + 1))
plt.show()

# Output expected mean and variance
expected_mean, expected_variance
```
